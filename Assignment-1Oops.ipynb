```python
"""Bank Account Create a class representing a bank account with attributes like account number, account holder name, and balance.
Implement methods to deposit and withdraw money from the account."""
class BankAccount:
    def __init__(self, account_number, account_holder_name, balance=0):
        self.account_number = account_number
        self.account_holder_name = account_holder_name
        self.balance = balance

    def deposit(self, amount):
        if amount > 0:
            self.balance += amount
            print(f"Deposited ₹{amount}. New balance: ₹{self.balance}")
        else:
            print("Invalid deposit amount. Please enter a positive amount.")

    def withdraw(self, amount):
        if amount > 0 and amount <= self.balance:
            self.balance -= amount
            print(f"Withdrew ₹{amount}. New balance: ₹{self.balance}")
        elif amount > self.balance:
            print("Insufficient funds.")
        else:
            print("Invalid withdrawal amount. Please enter a positive amount.")

account1 = BankAccount("1234567890", "John Doe")
account1.deposit(1000)
account1.withdraw(500)
account1.withdraw(1000)  
```

```python
""" Employee Management Create a class representing an employee with attributes like employee ID, name, and salary.
Implement methods to calculate the yearly bonus and display employee details."""
class Employee:
    def __init__(self, employee_id, name, salary):
        self.employee_id = employee_id
        self.name = name
        self.salary = salary

    def calculate_yearly_bonus(self, bonus_percentage):
        bonus = (self.salary * bonus_percentage) / 100
        return bonus

    def display_details(self):
        print("Employee Details:")
        print(f"Employee ID: {self.employee_id}")
        print(f"Name: {self.name}")
        print(f"Salary: ₹{self.salary}")

# Example usage
employee1 = Employee("EMP001", "Alice", 50000)
bonus_percentage = 10
bonus_amount = employee1.calculate_yearly_bonus(bonus_percentage)
print(f"Bonus Amount: ₹{bonus_amount}")
employee1.display_details()
```

```python
"""Vehicle Rental Create a class representing a vehicle rental system. Implement methods to rent a vehicle, return a vehicle, and display available vehicles."""
class Vehicle:
    def __init__(self, vehicle_id, vehicle_type, rental_rate_per_day):
        self.vehicle_id = vehicle_id
        self.vehicle_type = vehicle_type
        self.rental_rate_per_day = rental_rate_per_day
        self.is_rented = False

class VehicleRentalSystem:
    def __init__(self):
        self.available_vehicles = []

    def add_vehicle(self, vehicle):
        self.available_vehicles.append(vehicle)

    def rent_vehicle(self, vehicle_id):
        for vehicle in self.available_vehicles:
            if vehicle.vehicle_id == vehicle_id and not vehicle.is_rented:
                vehicle.is_rented = True
                return vehicle
        return None

    def return_vehicle(self, vehicle):
        vehicle.is_rented = False

    def display_available_vehicles(self):
        print("Available Vehicles:")
        for vehicle in self.available_vehicles:
            if not vehicle.is_rented:
                print(f"Vehicle ID: {vehicle.vehicle_id}")
                print(f"Vehicle Type: {vehicle.vehicle_type}")
                print(f"Rental Rate per Day: ₹{vehicle.rental_rate_per_day}")
                print("--------------------")

# Example usage
rental_system = VehicleRentalSystem()
car1 = Vehicle("CAR001", "Sedan", 500)
car2 = Vehicle("CAR002", "SUV", 800)
rental_system.add_vehicle(car1)
rental_system.add_vehicle(car2)
rental_system.display_available_vehicles()
rented_car = rental_system.rent_vehicle("CAR001")
if rented_car:
    print(f"Rented {rented_car.vehicle_type}")
else:
    print("Vehicle not available for rent.")
rental_system.return_vehicle(rented_car)
rental_system.display_available_vehicles()
```

```python
""" Library Catalog Create classes representing a library and a book. Implement methods to add books to the library, borrow books, and display available books."""
class Book:
    def __init__(self, title, author, isbn):
        self.title = title
        self.author = author
        self.isbn = isbn
        self.is_available = True

    def __str__(self):
        return f"{self.title} by {self.author} (ISBN: {self.isbn})"
    
    
class Library:
    def __init__(self):
        self.books = []

    def add_book(self, book):
        self.books.append(book)

    def borrow_book(self, isbn):
        for book in self.books:
            if book.isbn == isbn and book.is_available:
                book.is_available = False
                return book
        return None

    def display_available_books(self):
        print("Available Books:")
        for book in self.books:
            if book.is_available:
                print(book)
                
# Create a library
my_library = Library()

# Create books
book1 = Book("The Great Gatsby", "F. Scott Fitzgerald", "9780743273565")
book2 = Book("To Kill a Mockingbird", "Harper Lee", "9780060735336")

# Add books to the library
my_library.add_book(book1)
my_library.add_book(book2)

# Display available books
my_library.display_available_books()

# Borrow a book
borrowed_book = my_library.borrow_book("9780060735336")
if borrowed_book:
    print(f"Borrowed: {borrowed_book}")
else:
    print("Book not available or invalid ISBN")

# Display available books again
my_library.display_available_books()
```

```python
""" Product Inventory Create classes representing a product and an inventory system. Implement methods to add 
products to the inventory, update product quantity, and display available products."""
class Product:
    def __init__(self, product_id, name, price, quantity):
        self.product_id = product_id
        self.name = name
        self.price = price
        self.quantity = quantity

    def __str__(self):
        return f"Product ID: {self.product_id}\nName: {self.name}\nPrice: ₹{self.price}\nQuantity: {self.quantity}\n"

class InventorySystem:
    def __init__(self):
        self.products = []

    def add_product(self, product):
        if product not in self.products:
            self.products.append(product)
            print(f"Product '{product.name}' added to inventory.")
        else:
            print(f"Product '{product.name}' already exists in inventory.")

    def update_product_quantity(self, product_id, new_quantity):
        for product in self.products:
            if product.product_id == product_id:
                product.quantity = new_quantity
                print(f"Quantity of '{product.name}' updated to {new_quantity}.")
                return
        print(f"Product with ID '{product_id}' not found.")

    def display_available_products(self):
        print("Available Products:")
        for product in self.products:
            print(product)


inventory = InventorySystem()
product1 = Product("PROD001", "Laptop", 50000, 10)
product2 = Product("PROD002", "Smartphone", 30000, 20)
inventory.add_product(product1)
inventory.add_product(product2)
inventory.display_available_products()
inventory.update_product_quantity("PROD001", 5)
inventory.display_available_products()
inventory.add_product(product1)
```

```python
"""Shape Calculation Create a class representing a shape with attributes like length, width, and height. Implement methods to calculate the area and perimeter of the shape."""
class Shape:
    def __init__(self, length, width, height=0):
        self.length = length
        self.width = width
        self.height = height

    def calculate_area(self):
        raise NotImplementedError("Calculate area method not implemented")

    def calculate_perimeter(self):
        raise NotImplementedError("Calculate perimeter method not implemented")

class Rectangle(Shape):
    def calculate_area(self):
        return self.length * self.width

    def calculate_perimeter(self):
        return 2 * (self.length + self.width)

class Square(Shape):
    def __init__(self, side_length):
        super().__init__(side_length, side_length)

    def calculate_area(self):
        return self.length ** 2

    def calculate_perimeter(self):
        return 4 * self.length

class Cube(Shape):
    def calculate_area(self):
        return 6 * (self.length * self.width)

    def calculate_perimeter(self):
        return 12 * self.length

# Example usage
rectangle = Rectangle(5, 3)
square = Square(4)
cube = Cube(2, 2, 2)
print("Rectangle Area:", rectangle.calculate_area())
print("Rectangle Perimeter:", rectangle.calculate_perimeter())

print("Square Area:", square.calculate_area())
print("Square Perimeter:", square.calculate_perimeter())

print("Cube Area:", cube.calculate_area())
print("Cube Perimeter:", cube.calculate_perimeter())
```

```python
""" Student Management Create a class representing a student with attributes like student ID, name, and grades.
Implement methods to calculate the average grade and display student details."""
class Student:
    def __init__(self, student_id, name, grades):
        self.student_id = student_id
        self.name = name
        self.grades = grades

    def calculate_average_grade(self):
        if self.grades:
            return sum(self.grades) / len(self.grades)
        else:
            return 0

    def display_details(self):
        print("Student Details:")
        print(f"Student ID: {self.student_id}")
        print(f"Name: {self.name}")
        print(f"Grades: {self.grades}")
        print(f"Average Grade: {self.calculate_average_grade()}")

# Example usage
student1 = Student("STU001", "Alice", [85, 90, 78])
student2 = Student("STU002", "Bob", [])

student1.display_details()
student2.display_details()
```

```python
""" Social Media Profile Create a class representing a social media profile with attributes like username and posts. 
Implement methods to add posts, display posts, and search for posts by keyword."""
class SocialMediaProfile:
    def __init__(self, username):
        self.username = username
        self.posts = []

    def add_post(self, post):
        self.posts.append(post)

    def display_posts(self):
        for post in self.posts:
            print(post)

    def search_posts(self, keyword):
        matching_posts = [post for post in self.posts if keyword in post.lower()]
        return matching_posts

# Example usage
profile = SocialMediaProfile("user123")
profile.add_post("Hello everyone!")
profile.add_post("This is a second post.")
profile.add_post("Searching for posts containing 'hello'.")

profile.display_posts()

search_results = profile.search_posts("hello")
print("Search results:")
for post in search_results:
    print(post)
```

```python
""" ToDo List Create a class representing a ToDo list with attributes like tasks and due dates. 
Implement methods to add tasks, mark tasks as completed, and display pending tasks."""
class Task:
    def __init__(self, task, due_date):
        self.task = task
        self.due_date = due_date
        self.completed = False

    def __str__(self):
        return f"{self.task} (Due: {self.due_date})"

class ToDoList:
    def __init__(self):
        self.tasks = []

    def add_task(self, task):
        self.tasks.append(task)

    def mark_task_completed(self, task_index):
        if 0 <= task_index < len(self.tasks):
            self.tasks[task_index].completed = True

    def display_pending_tasks(self):
        print("Pending Tasks:")
        for index, task in enumerate(self.tasks):
            if not task.completed:
                print(f"{index + 1}. {task}")

todo_list = ToDoList()
todo_list.add_task(Task("Complete project report", "2024-09-10"))
todo_list.add_task(Task("Attend meeting", "2024-09-05"))
todo_list.add_task(Task("Buy groceries", "2024-09-07"))

todo_list.display_pending_tasks()

todo_list.mark_task_completed(1)
todo_list.display_pending_tasks()
```
